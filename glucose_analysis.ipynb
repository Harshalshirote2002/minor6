```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import os
os.chdir('big-ideas-lab-glycemic-variability-and-wearable-device-data-1.1.0/')
```

```python
glucose_vals=[]
for filename in os.listdir():
    if os.path.isdir(filename):
        glucose_vals.append(pd.read_csv(filename +'/Dexcom_' + filename + '.csv')['Glucose Value (mg/dL)'])
```

```python
for i in range(len(glucose_vals)):
    glucose_vals[i]=glucose_vals[i][12:]
```

```python
plt.figure(figsize=(25,10))
plt.scatter(np.arange(len(glucose_vals[2])), glucose_vals[2], marker='X', linewidths=0.01)
```

```python
sum=0
for i in range(len(glucose_vals)):
    print((glucose_vals[i]>180).sum())
    sum+=(glucose_vals[i]>180).sum()
print(f'sum of all={sum}')
```

```python
sum=0
for i in range(len(glucose_vals)):
    print((115<glucose_vals[i]).sum())
    sum+=(115<glucose_vals[i]).sum()
print(f'sum of all={sum}')
```

```python
values=np.array(glucose_vals[0])
for i in range(1,16):
    values= np.concatenate((values, np.array(glucose_vals[i])))
```

```python
plt.figure(figsize=(20,6))
plt.plot(values)
```
